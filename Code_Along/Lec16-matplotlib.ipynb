```python
import matplotlib.pyplot as plt
import numpy as np

# array from -10 to 10 with numbers equally spaced between
x = np.linspace(-10, 10, 20)  # X globala namespace, kan heta vadsom

f = lambda x: np.cos(x)        # X locala namespace, skriver inte över x på rad 5
# equviliant of lambda
# def f(x):
#    return npcos(x)
```

## MATLAB Style Interface

```python
plt.plot(x, f(x), 'o--') # - , . 
```

```python
x = np.linspace(10,-10, 100)

plt.figure(figsize=(8, 4), dpi = 200)  # storlek, dots per inches
plt.plot(x, f(x))           # plot 
plt.title("cos(x)")         # title
plt.xlabel("x")             # x corner label
plt.ylabel("y")             # y corner label
# plt
```

## OOP Approach

```python
help(plt) # hover and click ctrl
```

```python
# instantiated fig, ax from figure amd axes

fig, ax = plt.figure(figsize = (8,4)), plt.axes()
ax.plot(x, f(x)) # ax object has 
ax.set(xlabel = "x", ylabel="y", title = "$\cos{x}$", xlim = (-5,5), ylim = (-1.2,1.2))
fig.savefig("../assets/cos(x).png", dpi=200) # exporterar grafen
```

## Subplots

```python
# MATLAB style (räknar ifrån 1 inte 0)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1) # (rows, columns, panelnumer)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))

plt.subplot(1,2,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trinonometric functions")
```

```python
# MATLAB style (räknar ifrån 1 inte 0)
plt.figure(figsize=(8,4))
plt.subplot(2,1,1) # (rows, columns, panelnumer)
plt.plot(x, f(x))
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, f(x))
plt.title("cos(x)")

plt.suptitle("Trinonometric functions")
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,5))

ax[0].plot(x, f(x))
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
```
